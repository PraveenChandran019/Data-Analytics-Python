```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
#Resize your graph
plt.figure(figsize=(5,3), dpi=150)

#plt.plot(x,y, label = "2x", color='y',linewidth =5, marker='.', linestyle='--' ,markersize =  10, markeredgecolor='blue' )
#use short hand notation
#fmt = '[color][marker][line]'
plt.plot(x,y,'b.-', label = "2x" )
#line number 2
x2 = np.arange(0,4.5,0.5)
print(x2)
plt.plot(x2[:6],x2[:6]**2,x2,x2**2,'r.--', label = 'x**2')
plt.plot(x2[5:],x2[5:]**2,'r.-')




plt.title("Our First Plot", fontdict={"fontname": "comic sans ms",'fontsize':20})
plt.xlabel("X axis")
plt.ylabel("y axis")

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
plt.savefig("mygraph.png",dpi=150)
```

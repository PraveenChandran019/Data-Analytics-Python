```python
def contract_info(contract):
    if contract == "Free" or "On Loan" in contract:
         start_date = np.nan
         end_date = np.nan
         contract_length = np.nan
    else:
        start_date , end_date = contract.split(" ~ ")
        start_year = int(start_date[:4])
        end_year = int(end_date[:4])
        contract_length = int(start_year - end_year)
        return start_date, end_date, contract_length
        
```

```python
new_cols = ['Start Year', 'End Year', 'Contract Year']
new_data = fifa["Contract"].apply(lambda x: pd.Series(contract_info(x)))
for i in (range(len(new_cols))):
    fifa.insert(loc = fifa.columns.get_loc("Contract")+1+i, column=new_cols[i], value=new_data[i])
```

```python
# contract categories

def contract_categorize(contract):
    if contract == "Free":
        return "Free"
    elif "On Loan" in contract:
        return "On Loan"
    else:
        return "Contract"
```

```python
contract_values = fifa["Contract"].apply(lambda x: contract_categorize(x))
fifa.insert(loc = fifa.columns.get_loc("Contract Year"), column = "Contract status", value = contract_values)
```

```python
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inches = height.split("'")
        total_inches = int(feet)* 12 + int(inches.strip('"'))
        return round(total_inches * 2.54)
        
```

```python
fifa["Height"] = fifa["Height"].apply(lambda x: convert_height(x))
```
